**IMPORTING LIBRARIES**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**READING DATA**

```python
df=pd.read_csv("/content/creditcard_data.csv")
df.head(5)
```

Here the dataset is unbalanced highly , and then here:

*   0 - Normal Transaction
*   1 - Fraudulent Transaction



```python
df.describe()
```

```python
# Seperating the data:
legit = df[df.Class==0]
fraud = df[df["Class"] == 1]
```

```python
print(fraud.shape)
```

```python
print(legit.shape)
```

```python
legit_sample = legit.sample(n=492)
df=pd.concat([legit_sample,fraud],axis=0)
```

```python
df["Class"].value_counts()
```

```python
fraud.describe()
```

```python
# Comparing both the values:
df.groupby("Class").mean()
```

Under-Sampling -> I am building a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

**SPLITTING & MODEL TRAINING**

```python
X = df.drop("Class",axis=1)
Y = df["Class"]
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data

model.fit(X_train, Y_train)
```

**MODEL EVALUATION**

```python
# Accuracy on Training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# Accuracy on Testing Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
